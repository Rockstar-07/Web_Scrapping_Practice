```python
import requests
from bs4 import BeautifulSoup
url = "https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_population"
page = requests.get(url)
page.content
```

```python
soup = BeautifulSoup(page.content,'html.parser')
soup
```

```python
data=soup.find(class_="wikitable")
data
```

```python
head=[t.get_text().strip() for t in data.find_all('th')[:12]]
head
```

```python
remaining=[t.get_text().strip() for t in data.find_all('td')]
remaining
```

```python
d={}
for k in head:
    d[k]=[]
    
for i in range(0,len(remaining),12):
    d['Rank'].append(remaining[i])
    d['State or union territory'].append(remaining[i+1])
    d['Population'].append(remaining[i+2])
    d['Population (%)'].append(remaining[i+3])
    d['Decadal growth(20012012)'].append(remaining[i+4])
    d['Rural population'].append(remaining[i+5])
    d['Percent rural'].append(remaining[i+6])
    d['Urban population'].append(remaining[i+7])
    d['Percent urban'].append(remaining[i+8])
    d['Area[16]'].append(remaining[i+9])
    d['Density[a]'].append(remaining[i+10])
    d['Sex ratio'].append(remaining[i+11])
```

```python
import pandas as pd
df=pd.DataFrame(d)
df
```
